Spotify is a music streaming app that allows its users to access a variety of songs and podcasts. This dataset covers the information and features of 2000 songs between the years 1998 through 2020. I chose this dataset because I wanted to have an understanding of what attributes or features makes a song popular.

```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import pandas as pd
```

```python
df.drop_duplicates(inplace=True)
```

```python
print("✅ Duplicates removed. New shape of the DataFrame:", df.shape)
```

```python
df = df.drop_duplicates()
df = df.dropna(how='any')
df.columns = df.columns.str.lower().str.replace(' ', '_')
```

```python
plt.style.use('ggplot')
```

```python
plt.bar(df['year'].value_counts().sort_index().index, # This accesses the years and make sure they are displayed in ascending order
        df['year'].value_counts().sort_index().values, color='pink') # This access the count of the songs by year and makes sure they are displayed in ascending order

plt.xlabel('Year')
plt.ylabel('Count of Songs')
plt.title('Number of Songs by Year')
plt.savefig('plot1.png')
plt.show()
```

In this barplot the insight we can see is that the distribution is somewhat evenly spread between the years 1998 and 2020. This indictates that Spotify's music offerings for streaming caters  to a variety of audiences providing a more comprehensive library that users can choose from. The year with the highest number of songs was 2013 with some high peaks in other years possibly indicating that there may have been years where songs were not as popular or released as frequently as other years. It is also important to note that there is data missing from 2021-2024 meaning that there could be some data missing to inform on current song popularity trends.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Calculate the top 10 artists by song count
top_artists = df['artist'].value_counts().nlargest(10)

# Create a DataFrame for plotting
top_artists_df = top_artists.reset_index()
top_artists_df.columns = ['artist', 'count']

# Plot the bar chart with hue assigned to 'artist'
plt.figure(figsize=(10, 6))
sns.barplot(data=top_artists_df, x='count', y='artist', hue='artist', palette='viridis', dodge=False, legend=False)
plt.title("Top 10 Artists by Number of Songs")
plt.xlabel("Number of Songs")
plt.ylabel("Artist")
plt.tight_layout()
plt.savefig("top_artists.png", dpi=300, bbox_inches='tight')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.lineplot(data=df, x='year', y='danceability', estimator='mean')
plt.title("Average Danceability Over Time")
plt.ylabel("Danceability")
plt.xlabel("Year")
plt.tight_layout()
plt.savefig("spotifyplaylist.csv")
plt.close()
```

```python
plt.scatter(df['loudness'], df['popularity'], color= 'purple', alpha=0.2)
plt.xlabel('Loudness')
plt.ylabel('Popularity')
plt.title('Loudness vs Popularity')
plt.savefig('plot2.png')
plt.show()
```

In this Scatterplot, the insight we can see there's a positive correlation between loudness and popularity indicating that loudness ina song does increase its popularity. It is important to note that there is a cluster of outliers at the bottom of the graph, indicating that while loudness has an influence on song popularity, there can be other factors that also affect song popularity.

 I am choosing to analyze the relationship between valence and popularity and visualize it in a scatterplot as it would be the best choice to show if there is a correlation between two numerical variables.

```python
plt.scatter(df['valence'], df['popularity'], color='orange', alpha=0.2)
plt.xlabel('Valence')
plt.ylabel('Popularity')
plt.title('Valence vs Popularity')
plt.savefig('plot5.png')
plt.show()
```

```python
import plotly.express as px

fig = px.scatter(df, x='energy', y='popularity',
                 color='genre' if 'genre' in df.columns else None,
                 hover_data=['name', 'artist_name'] if 'name' in df.columns else None,
                 title='Popularity vs Energy (Interactive)')
fig.show()
```

```python
df['popularity_tier'] = pd.cut(df['popularity'], bins=[0, 50, 70, 100], labels=['Low', 'Medium', 'High'])

fig = px.histogram(
    df,
    x='danceability',
    color='popularity_tier',
    nbins=30,
    title='Danceability vs Popularity Tier (Interactive)',
    labels={'danceability': 'Danceability Score', 'popularity_tier': 'Popularity Level'},
    template='plotly_white'
)
fig.update_layout(bargap=0.1)
fig.show()
```

```python
import plotly.express as px
fig = px.box(df,
             x='genre',
             y='popularity',
             points="all",
             color='genre',
             title='Distribution of Popularity by Genre',
             labels={"popularity": "Popularity Score", "genre": "Genre"},
             template="plotly_white")

fig.show()
```

```python
columns = [
    'duration_ms', 'explicit', 'year', 'popularity', 'danceability',
    'energy', 'key', 'loudness', 'mode', 'speechiness', 'acousticness',
    'instrumentalness', 'liveness', 'valence', 'tempo'
]

# Create a dummy DataFrame to replicate structure
np.random.seed(42)
df = pd.DataFrame(np.random.rand(2000, len(columns)), columns=columns)

# Generate and display correlation heatmap
corr_matrix = df.corr()

plt.figure(figsize=(12, 10))
sns.heatmap(corr_matrix, annot=True, fmt=".2f", cmap='Spectral', center=0, square=True, linewidths=0.5, linecolor='black')
plt.title("Enhanced Correlation Heatmap of Numeric Features", fontsize=16)
plt.tight_layout()
plt.show()
```

```python
fig2 = px.scatter(
    df,
    x='year',
    y='popularity',
    trendline='ols',
    color='popularity',
    color_continuous_scale='Plasma',
    title='📅 Popularity vs Year (Interactive with Trend Line)',
    hover_data=['danceability', 'energy']
)

fig2.update_layout(width=800, height=500)
fig2.show()
```

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(data=df, x='valence', y='danceability', hue='popularity', palette='viridis')
plt.title("🎵 Valence vs Danceability (Colored by Popularity)")
plt.xlabel("Valence (Musical Positivity)")
plt.ylabel("Danceability (Rhythmic Appeal)")
plt.legend(title='Popularity', loc='best')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
fig3 = px.scatter(
    df,
    x='valence',
    y='danceability',
    color='popularity',
    color_continuous_scale='Turbo',
    title='🎵 Valence vs Danceability (Colored by Popularity)',
    hover_data=['year', 'energy', 'liveness'],
    template='plotly_white'
)

fig3.update_traces(marker=dict(size=7, opacity=0.6, line=dict(width=0.5, color='DarkSlateGrey')))
fig3.update_layout(width=800, height=500)
fig3.show()
```
